### 卡方分佈

```python
#自由度相差1，把卡方化成藝術品
import numpy as np
import matplotlib.pyplot as plt

xlim = [0, 50]
x = np.linspace(xlim[0], xlim[1], 1000)
df = np.arange(4, 32, 1)
y = chi2.pdf(x.reshape(-1,1), df)
plt.figure()
plt.axis([xlim[0], xlim[1],0,0.2])#座標範圍設定
plt.plot(x, y, lw=1.5, color='skyblue')
plt.title(r'$\chi^2$ Distribution')
plt.yticks([0, 0.1, 0.2])
plt.show()
```

### T分配

```python
import numpy as np
from scipy.stats import t
from scipy.stats import norm
import matplotlib.pyplot as plt

xlim = [-5, 5]
x = np.linspace(xlim[0], xlim[1], 100)
df1=np.arange(0.1, 1, 0.1)
df2=np.arange(3, 30, 3)
df = np.hstack((df1, df2)) 
y = t.pdf(x.reshape(-1,1), df)

mu, sigma = 0, 1
x2 = np.linspace(xlim[0], xlim[1], 1000)
y2 = norm.pdf(x2, mu, sigma)

plt.figure()
plt.plot(x, y, lw=1, color='blue', alpha=0.5)
plt.plot(x2, y2,color='red',label='Z distribution')
plt.title('T Distribution')
plt.legend()
plt.show()
```
