### 1.$f(x) = \sin(x) + \cos(x)  $

```python
# 第一題 週期做多一點   顏色
x = np.linspace(-3*np.pi, 3*np.pi, 200)
def f(x): return np.sin(x) + np.cos(x)
def f2(x): return np.cos(x)


def f3(x): return np.sin(x)


fig = plt.figure(figsize=[6, 4])  # 圖形長寬
plt.plot(x, f(x), linestyle='-', label='sin(x)+cos(x)',
         linewidth=2, color='royalblue')
plt.plot(x, f2(x), linestyle='dotted',
         label='cos(x)', linewidth=1.5, color='gray')
plt.plot(x, f3(x), linestyle='dotted', label='sin(x)',
         linewidth=1.5, color='chocolate')

plt.legend()
plt.grid(visible=True, linestyle='-', linewidth=0.5)
plt.xlabel('X'), plt.ylabel('f(x)')
plt.ylim(-3, 3)
plt.xticks([-3*np.pi, -2*np.pi, -np.pi, 0, np.pi, 2*np.pi, 3*np.pi],
           ['-3$\pi$', '-2$\pi$', '-$\pi$', '0', '$\pi$', '2$\pi$', '3$\pi$'])
plt.title('$f(x) = sin(x) + cos(x)$')
plt.savefig(img_dir + 'hw2-1.pdf', format='pdf')
plt.show()
```

### 3.$f(x) = \sqrt[3]{\frac{4-x^3}{1+x^2}} $

```python
# 第三題
x = np.arange(-10, 10, 0.001)
y = np.cbrt((4-x**3)/(1+x**2))

p1 = [-1.512745, 0]
p2 = [1.314071, 1.587401]


fig = plt.figure(figsize=[6, 4])
ax = plt.gca()
plt.scatter(p1, p2)
plt.plot(x, y)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('f(x)')
plt.title(r'$f(x) = \sqrt[3]{(\frac{4-x^3}{1+x^2})}$')
ax.annotate('relative minimum',
            xy=(-1.512745, 1.2), xytext=(-7, 0.1),
            arrowprops=dict(arrowstyle='->',
                            connectionstyle='arc3, rad=0.3'), fontsize=12)
ax.annotate('relative maximum',
            xy=(0.2, 1.587401), xytext=(3, 1.1),
            arrowprops=dict(arrowstyle='->',
                            connectionstyle='arc3, rad=0.3'), fontsize=12)
plt.savefig(img_dir + 'hw2-3.pdf', format='pdf')
plt.show()
```

### 5.$f(x) = \frac{1}{2\sqrt{2\pi}} e^{-\frac{(x-1)^2}{8}} $

```python
# 範例五
x = np.linspace(-8, 10, 100)
y = 1/(2*(2*np.pi)**(1/2)) * np.exp(-(x-1)**2/8)

fig = plt.figure(figsize=[6, 4])
fig.patch.set_facecolor('g')
fig.patch.set_alpha(0.2)
plt.plot([-20, 20], [0, 0], color='gray', linestyle='-.')
ax = plt.gca()
ax.set_ylim([-0.05, 0.25])
ax.plot(x, y, linewidth=2, color='skyblue', linestyle='-')
ax.grid(True)
ax.set_xlabel('X'), ax.set_ylabel('f(x)')
plt.xlim(-6, 8)
plt.scatter(1, 0.1999, color='skyblue')
ax.set_title(r'$f(x)=\frac{1}{2\sqrt{2\pi}}e^{-\frac{(x-1)^2}{8}}$')
ax.text(-3.8, -0.02, '$y=0$')
plt.savefig(img_dir + 'hw2-5.pdf', format='pdf')
plt.show()
```

### 第二小題 想法一
 
證明 $\gamma_n =S_n - ln(n+1)$

粉紅色長條面積 b

$b = 1*1 + (1/2)*1 + (1/3)*1 + ......+ (1/n)*1$ 

$\ \ \ = S_n$

函數 $f(x) = 1/x$ 底下的面積 r , $x>1$

if $x = 1$, $r = 0$

if $x = 2$, $r = \int_{1}^{3} \frac{1}{x}dx = ln3$

if $x = 3$, $r = \int_{1}^{4} \frac{1}{x}dx = ln4$

if $x = n$, $r = \int_{1}^{n+1} \frac{1}{x}dx = ln(n+1)$

$\gamma_n = b - r = S_n - ln(n+1)$



### 第二小題 想法二

$\gamma_n=\sum_{k=1}^n\int_{x=k}^{k+1} (\frac{1}{k}-\frac{1}{x})dx$

$\quad \; =\sum_{k=1}^n(\int_{k}^{k+1} \frac{1}{k}dx-\int_{k}^{k+1}\frac{1}{x}dx )$

$\quad \; =\sum_{k=1}^n(\frac{k+1}{k}-1-(ln|k+1|-ln|k|)$

$\quad \; =\sum_{k=1}^n(\frac{1}{k}+ln|k|- ln|k+1|)$

$\quad \; =\sum_{k=1}^n\frac{1}{k}+\sum_{k=1}^n(ln|k|-ln|k+1|)$

$\quad \; =S_n+ln1-ln2+ln2-ln3 \dots+ln(n)-ln(n+1)$

$\quad \;=S_n-ln(n+1)$

```python
y_locator = MultipleLocator(0.1)

n = []
Sn = []
for i in range(101):
    n.append(i)
    b = np.arange(1, n[i]+1)
    sn = (1/b).sum()
    Sn.append(sn)
# print(n)
# print(Sn)


x1 = np.arange(0, 101, 1)
y1 = np.log(x1+1)
result = Sn - y1
# print(result)

y2 = (1/2)*(1-(1/(x1+1)))

fig = plt.figure(figsize=[6, 4])
ax = plt.gca()
plt.plot([0, 100], [1, 1], label='1', linewidth=2, linestyle='-')
ax.plot(x1, result, label=r'$\gamma_n$', color='orange')
ax.plot(x1, y2, label=r'$\frac{1}{2}(1-\frac{1}{n+1})$', color='g')
ax.yaxis.set_major_locator(y_locator)
plt.legend()
plt.ylim(0.2, 1.1)
plt.grid(True)
plt.savefig(img_dir + 'hw2-14.pdf', format='pdf')
plt.show()
```
