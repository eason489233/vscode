### Start to learn matplotlib
- make a plot for function $f(x) = x^2 + 3x +5,\;\; -5 \leq x \leq 5$
- do basic computations 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 範例一
n = 100  # 用幾個點畫出圖型
x = np.linspace(-5, 5, n)  # x的範圍（-5到5）
y = x**2 + 3*x + 5  # 方程式

plt.plot(x, y)
plt.grid(True)  # 畫出背景方格線
plt.xlabel('X')
plt.ylabel('Y')
plt.title('$f(x)=x^2+3x+5$')
plt.show()  # just for py file
```

```python
# 範例二
x = np.arange(0, 7, 0.01)  # 第三個數字是說每0.01的位置就設一個點
def f(x): return x**3 - 10*x**2 + 29*x - 20


fig = plt.figure(figsize=[6, 4])  # 圖形長寬
plt.plot(x, f(x), linestyle='--', color='r')
plt.grid(visible=True, color='g', linestyle='-', linewidth=0.5)
plt.xlabel('X'), plt.ylabel('Y')
plt.title('$f(x)=x^3-10x^2+29x-20$')
plt.show()
```

```python
# 範例三
xmin, xmax = -1, 5
x = np.linspace(xmin, xmax, 100)
p = [1, -8, 16, -2, 8]
def f(x): return np.polyval(p, x)


fig = plt.figure()
ax = plt.gca()
ax.plot(x, f(x), linewidth=3, color='g', linestyle='-.')
ax.grid(True)
ax.set_xlabel('X'), ax.set_ylabel('f(x)')
ax.set_title('The range of $x$ is not appropriate')
ax.text(0, 25, 'Re-define the range of $x$')
# save the current figure as an eps file
plt.savefig('poly.eps', format='eps')
plt.show()
```

```python
# 範例3-2
x = np.linspace(-3*np.pi, 3*np.pi, 200)
#y=4/np.pi * np.sin(x)

y = 0
for i in range(50):
    k = 2*i + 1
    y = y+np.sin(k*x)/k
y = y * 4/np.pi
#y = 4 / np.pi * np.sin(x)
fig, ax = plt.subplots(1)
ax.plot(x, y, color='b')
ax.set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi)
ax.set_xticklabels(['-3$\pi$', '-2$\pi$', '$\pi$', '0', '$\pi$', '2$\pi$', '3$\pi$'],
                   fontsize=10, color='b')
ax.set_ylim([-3, 3])
ax.grid(True)
plt.show()
```

```python
# 範例4
n = [1, 2, 3]
colors = ['r', 'g', 'b']
x = np.linspace(-3*np.pi, 3*np.pi, 200)
for i in n:
    y = np.sin(i*x)
    plt.plot(x, y, label='sin({}x)'.format(i),
             color=colors[i - 1])

plt.legend()
plt.ylim(-2.5, 2.5), plt.grid(True), plt.xlabel('X')
plt.title('Demonstration of legend for each line')
plt.show()
```

```python
# 範例五
import matplotlib.patches as pat

fig, ax = plt.subplots(figsize=(4, 3))
rec1 = pat.Rectangle((1, 1), 2, 3, color='green', alpha=0.5)
ax.add_patch(rec1)
rec2 = pat.Rectangle((2, 2), 2, 3, color='red', alpha=0.5)
ax.add_patch(rec2)
plt.xlim([0, 5]), plt.ylim([0, 6])  # x,y的範圍
plt.show()
```

```python
# 範例六
x = np.linspace(-0.5, 1.6, 100)
def f(x): return 2*x - x**2
def g(x): return x**2


plt.plot(x, f(x), x, g(x))  # draw two lines in a command
plt.fill_between(x, f(x), g(x), where=f(x) > g(x),
                 color='pink')  # defaut : y1>y2
ax = plt.gca()
# Re-arrange four axes spines
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['top'].set_visible(False)  # take off top line
ax.spines['right'].set_visible(False)
ax.set_facecolor('#D5FAF9')
ax.annotate('$\int_0^1 (f(x) - g(x)) \; dx$',
            xy=(0.5, 0.5), xytext=(0.3, 1.5),
            arrowprops=dict(arrowstyle='->',
                            connectionstyle='arc3, rad=0.3'), fontsize=14)
plt.show()
```
