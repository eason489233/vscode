### 卡方分配

```python
# 自由度相差1，把卡方化成藝術品
from scipy.stats import chi2
xlim = [0, 50]
x = np.linspace(xlim[0], xlim[1], 1000)
df = np.arange(4, 32, 1)
y = chi2.pdf(x.reshape(-1, 1), df)

fig = plt.figure(figsize=[6, 4])
plt.style.use('fivethirtyeight')
plt.axis([xlim[0], xlim[1], 0, 0.2])  # 座標範圍設定
plt.plot(x, y, lw=1.5, color='skyblue')
plt.title(r'$\chi^2$ Distribution')
plt.grid(True)
plt.yticks([0, 0.1, 0.2])
plt.savefig(img_dir+'chi_distribution.pdf', format='pdf')
plt.show()
```

### T分配 自由度兩組，(0.1,1,0.1)和(3,30,3)

```python
from scipy.stats import t
from scipy.stats import norm

xlim = [-5, 5]
x = np.linspace(xlim[0], xlim[1], 100)
df1 = np.arange(0.1, 1, 0.1)
df2 = np.arange(3, 30, 3)
df = np.hstack((df1, df2))
y = t.pdf(x.reshape(-1, 1), df)

mu, sigma = 0, 1
x2 = np.linspace(xlim[0], xlim[1], 100)
y2 = norm.pdf(x2, mu, sigma)

fig = plt.figure(figsize=[6, 4])
plt.plot(x, y, lw=1.5, color='orange', alpha=0.5)
plt.plot(x2, y2, lw=2, color='red', alpha=0.8, label='Z distribution')
plt.title('T Distribution')
plt.legend()
plt.savefig(img_dir+'t_distribution.pdf', format='pdf')
plt.show()
```

### Beta分配

```python
from scipy.stats import beta

a = 9
b = np.arange(1, 30)
x = np.linspace(0, 1, 200)
y = beta.pdf(x.reshape(-1, 1), a, b)
fig = plt.figure(figsize=[6, 4])
plt.plot(x, y, c='g', lw=2, alpha=0.7)
plt.title(r'$\beta$ Distribution')
plt.savefig(img_dir+'beta_distribution.pdf', format='pdf')
plt.show()
```

### boxplot

```python
from scipy.stats import norm
rng = np.random.default_rng(seed=10)
n = 1000
x1 = norm.rvs(size=n, random_state=rng)  # N(0,1)
x2 = norm.rvs(loc=-1, scale=2, size=n)  # N(1,2)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 6))
ax1.boxplot(np.c_[x1, x2], notch=True, vert=True, labels=['X1', 'X2'])

boxprops = dict(linestyle='--', linewidth=3, color='darkgoldenrod')
flierprops = dict(marker='o', markerfacecolor='black',
                  markersize=8, linestyle='none')  # define outliers
labels = ['N(0,1)', 'N(1,4)']
ax2.boxplot(np.c_[x1, x2], boxprops=boxprops,
            flierprops=flierprops, labels=labels)
plt.savefig(img_dir+'boxplot.pdf', format='pdf')
```

### ECDF 用指數分配

```python
from scipy.stats import expon
n = 1000
x = expon.rvs(scale=0.2, size=n)
x_sort = np.sort(x)
F = np.arange(1, n+1)/n

# 母體資訊
X = np.linspace(x_sort[0], x_sort[-1], 1000)
Y = expon.cdf(X, scale=0.2)
fig = plt.figure(figsize=[6, 4])
plt.plot(X, Y, linestyle='-', label='real CDF', alpha=0.6)
plt.plot(x_sort, F, linestyle='--', label='Empirial CDF')
plt.legend()
plt.title('ECDF of Exponentioal Distribution')
plt.savefig(img_dir+'ecdf_exp.pdf', format='pdf')
plt.legend()
```

### x服從gamma(n,lambda)

```python
from scipy.stats import gamma
n = 1000
beta = (1/5)
x = gamma.rvs(10, scale=beta, size=n)
y = ((2 * x)/beta)
fig = plt.figure(figsize=[6, 4])
plt.hist(y, bins=100, density=True, alpha=0.5)

y_sort = np.sort(y)
x2 = np.linspace(0, 50, 1000)
y2 = chi2.pdf(x2, 20)
plt.title('Sampling Ditribution of $Y=2 \lambda X$,n=1000')
plt.plot(x2, y2, lw=1.5, color='red')
plt.savefig(img_dir+'sample_gamma_to_chi.pdf', format='pdf')
```
